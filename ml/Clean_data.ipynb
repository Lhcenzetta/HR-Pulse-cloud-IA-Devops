```python
import pandas as pd
import os
from dotenv import load_dotenv

load_dotenv()
path_data = os.getenv("path_data")
df = pd.read_csv(path_data)
df
```

```python
df.info()
```

```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df
```

```python
text_cols = ["job_title", "job_description", "company_name", "location", "headquarters",
             "size", "type_of_ownership", "industry", "sector", "revenue", "competitors"]

for col in text_cols:
    df[col] = df[col].str.strip()
    df[col] = df[col].str.replace(r'\n', ' ', regex=True)  
df
```

```python
import re
def parse_salary(s):
    numbers = re.findall(r'\d+', str(s))
    if len(numbers) == 2:
        return (int(numbers[0]) + int(numbers[1])) / 2 * 1000
    return None

df["salary_clean"] = df["salary_estimate"].apply(parse_salary)
```

```python
load_dotenv()
endpoint = os.getenv("endpoint")
api_key = os.getenv("api_key")
```

```python
from azure.ai.textanalytics import TextAnalyticsClient
from azure.core.credentials import AzureKeyCredential

client = TextAnalyticsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(api_key)
)
MAX_LENGTH = 5000 

def extract_skills(text):
    text = text[:MAX_LENGTH]
    response = client.recognize_entities([text])[0]
    skills = []
    for entity in response.entities:
        if entity.category in ["Skill", "Technology", "Organization"]:
            skills.append(entity.text)
    return list(set(skills))

df["skills_extracted"] = df["job_description"].apply(extract_skills)
```
